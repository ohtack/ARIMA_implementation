## 1st order differencing

```python
price = pd.DataFrame(data["Adj Close"])
price.columns = ['Adj Close']
price['1st diff'] = price['Adj Close'].diff(periods = 1)
price.head()
```

```python
price['1st diff'].dropna().plot(figsize = (15,7))
plt.xlabel("Date", fontsize=14)
plt.ylabel("1st Diffrencing Stock Price", fontsize=14)

plt.show()
```

```python
print("1st Diffrencing Stock Price")
print()
dfTest(price['1st diff'].dropna())
```

#####  Reject the null hypothesis (H0) and the data is stationary.

## 2nd order differencing

```python
price['2nd diff'] = price['Adj Close'].diff(periods = 1).diff()
price.head()
```

```python
price['2nd diff'].dropna().plot(figsize = (15,7))
plt.xlabel("Date", fontsize=14)
plt.ylabel("2nd Diffrencing Stock Price", fontsize=14)

plt.show()
```

```python
print("2nd Diffrencing Stock Price")
print()
dfTest(price['2nd diff'].dropna())
```

#####  Reject the null hypothesis (H0) and the data is stationary.
