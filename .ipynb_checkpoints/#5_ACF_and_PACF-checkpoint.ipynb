```python
from data_load import *
from dfTest import *
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
price = pd.DataFrame(data["Adj Close"])
# plot_acf(price["Adj Close"])
# plot_pacf(price["Adj Close"])

fig = plt.figure(figsize=(10, 10))

ax1 = fig.add_subplot(211)
fig = plot_acf(price["Adj Close"], ax=ax1,
               title="Autocorrelation on Original Stock Price") 

ax2 = fig.add_subplot(212)
fig = plot_pacf(price["Adj Close"], ax=ax2,
               title="Partial Autocorrelation on Original Stock Price")

plt.show()
```

```python
price['1st diff'] = price['Adj Close'].diff(periods = 1)

fig = plt.figure(figsize=(10, 10))

ax1 = fig.add_subplot(311)
fig = plot_acf(price["Adj Close"].dropna(), ax=ax1,
               title="Autocorrelation on Original Stock Price") 
ax2 = fig.add_subplot(312)
fig = plot_acf(price['1st diff'].dropna(), ax=ax2, 
               title="1st Order Differencing")
ax3 = fig.add_subplot(313)
fig = plot_acf(price['1st diff'].dropna(), ax=ax3, 
               title="2nd Order Differencing")
```

##### Original stock price ACF shows decaying very slowly, which is non-stationary moving.
##### On the other hand, 1st and 2nd order differencing shows exponential decay, which is stationary series.
