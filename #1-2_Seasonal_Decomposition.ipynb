```python
from data_load import *
from dfTest import *
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
price = pd.DataFrame(data["Adj Close"])
price.columns = ['Adj Close']
#price.index = pd.to_datetime(price.index)
decomp = seasonal_decompose(price['Adj Close'], model='additive', period = 20)
trend = decomp.trend
seasonal = decomp.seasonal
residual = decomp.resid
```

```python
plt.figure(figsize=(12,7))
plt.subplot(411)
plt.plot(price['Adj Close'], label='Original', color="blue")
plt.legend(loc='best')
plt.subplot(412)
plt.plot(trend, label='Trend', color="blue")
plt.legend(loc='best')
plt.subplot(413)
plt.plot(seasonal,label='Seasonality', color="blue")
plt.legend(loc='best')
plt.subplot(414)
plt.plot(residual, label='Residuals', color="blue")
plt.legend(loc='best')
plt.tight_layout()
```

```python
dfTest(residual.dropna())
```

```python
#price.index = pd.to_datetime(price.index)
price['1st diff'] = price['Adj Close'].diff(periods = 1)
decomp = seasonal_decompose(price['1st diff'].dropna(), model='additive', period = 20)
trend = decomp.trend
seasonal = decomp.seasonal
residual = decomp.resid

plt.figure(figsize=(12,7))
plt.subplot(411)
plt.plot(price['Adj Close'], label='Original', color="blue")
plt.legend(loc='best')
plt.subplot(412)
plt.plot(trend, label='Trend', color="blue")
plt.legend(loc='best')
plt.subplot(413)
plt.plot(seasonal,label='Seasonality', color="blue")
plt.legend(loc='best')
plt.subplot(414)
plt.plot(residual, label='Residuals', color="blue")
plt.legend(loc='best')
plt.tight_layout()
```
